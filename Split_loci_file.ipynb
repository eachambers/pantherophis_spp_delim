Below code is from Build_Nexus_file script, which calculates the # sites in each block (so the sequence length for each locus, in order of the `.loci` file)

```python
inputfile = "Pguttatus3.loci"
lookup = '//'
counting_lines = []
file = open(inputfile)

char_count = []

for line in file:
    char_count.append(len(line)-15) # 15 is the number of spaces allotted for taxon names

#print char_count

# line right before line containing locus number
file = open(inputfile)

with file as myFile:
    for num, line in enumerate(myFile, 1):
        if lookup in line:
            counting_lines.append(num-1)
            
#print counting_lines

loci_lengths = []
char_count[counting_lines[0]-1] # actual no. 3 in char_count because count starts at 0

for i in counting_lines:
    loci_lengths.append(char_count[i-1])

print loci_lengths
# can verify that we have all loci with:
#print len(loci_lengths)
```

```python
f = open("pantherophis_test.nexus", 'w')
f.write("#NEXUS\n\nbeginsets;\n")

with open('pantherophis_test.nexus', 'w') as f:
    for item in loci_lengths:
        f.write("\tcharset part %s\n" % item)
```

The following takes the `.loci` files output from iPyrad (recall: only our ddRAD datasets), splits the file by locus (split by //), and outputs each locus that is shared by at least two individuals into a separate `file*.txt` file in order to eventually count the number of shared loci between replicate samples.

```python
mod_file = "Pguttatus3.loci" # change to Pguttatus3_mod.loci if you want end format to be fasta
splittingtxt = '//'
filenameformat = 'pantherophis_locus#.phy'

def output(filenum,line):
    filename = filenameformat.replace('#',str(filenum) )
    fout=open(filename,'w')
    fout.write(line)
    fout.close

file = open(mod_file)
lines=file.read().split(splittingtxt)

for i in range(0,len(lines)):
    output(i+1, lines[i])
    
# FYI: this will make an extra locus (7022 in our case) with only a single line; delete this file manually.
```
